```python
#OPERACIONES BÁSICAS

# Vamos a generar varios ndarrays
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.random.rand(2, 3)

print("a = ")
print(a)

print("b = ")
print(b)

print("Suma de todos los números de b")
print(b.sum())
print(np.sum(b))

print("Media de cada columna de a")
print(a.mean(axis=0))
print(np.mean(a, axis=0))

print("Transpuesta de a")
print(a.transpose())
print(a.T)

print("10 * b = ")
print(10 * b)

print("a * b =")
print(a * b)

print("2 elevado a la matriz a")
print(np.power(2, a))

print("a elevada al cuadrado (elemento a elemento)")
print(np.power(a, 2))

print("Producto punto de a con b transpuesta")
print(a.dot(b.T))
print(np.dot(a, b.T))
print(a @ b.T)
```

```python
#Generamos un arreglo con 100 valores equiespaciados del seno desde 0 a 2$\pi$
a = np.sin(np.linspace(0, 2 * np.pi, 100))

#Lo convertimos en una matriz de 10 por *lo que sea* (-1), donde *lo que sea*
# es 10 en este caso,
print("a queda como: ")
print(a)
b = a.reshape((10, -1))
print("b queda como: ")
print(b)
print("donde b tiene ", b.ndim, " dimensiones, con una forma ", b.shape, "y con ", b.size, " elementos.")
```

```python
# Si queremos hacer que un vector se comporte como un vector renglon
a = np.arange(30).reshape(1,-1)
print("a = ", a)
print("a es de forma ", a.shape)

# Y si queremos que sea un vector columna hacemos esto
b = np.linspace(30, 35, 30).reshape(-1,1)
print("b = ", b)
print("b es de forma ", b.shape)

# Y para hacer una concatenacion de columnas entonces utilizamps la forma especial np._[]
c = np.c_[a.T, b]
print("c = ", c)
print("c es de forma ", c.shape)

#Y una concatenación de renglones es por lo tanto
#d = np.r_[a.T, b]
d = np.r_[a, b.T]
print("d = ", d)
print("d es de forma ", d.shape)
```

```python
#En la primer columna los valores entre -1 y 1, equiespaciados
col1 = np.linspace(-1,1,100).reshape(-1,1)
print(col1)
```

```python
# En la cuarta columna 1 si el valor de la segunda columna es
# mayor que cero y -1 en otro caso (revisa la función np.where)
col4 = np.where(col2 > 0, 1,-1)
print(col4)
```

```python
from matplotlib import ticker, cm

N = 100
x = np.linspace(-3.0, 3.0, N)
y = np.linspace(-2.0, 2.0, N)

X, Y = np.meshgrid(x, y)

# Vamos a crear un valor de Z usando la suma de dos funciones
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X * 10)**2 - (Y * 10)**2)
z = Z1 + 50 * Z2

# ponemos algunos puntos negativos para ver los problemas que pueden causar:
z[:5, :5] = -1

# Esta linea es para evitar un warning, pero se puede quitar para ver
# que avisa el sistema.
z = np.ma.masked_where(z <= 0, z)


# Automatic selection of levels works; setting the
# log locator tells contourf to use a log scale:
fig, ax = plt.subplots()
cs = ax.contourf(X, Y, z, locator=ticker.LogLocator(), cmap=cm.PuBu_r)

# Tambien se pueden ajustar los niveles de forma manual
# levs = np.linspace(0, z.max(), 200)
# cs = ax.contourf(X, Y, z, levs)

cbar = fig.colorbar(cs)

plt.show()
```

```python
#Genera un vector de datos de entrada x = np.linspace(0, 1, 1000)
#y grafica  sin(2πx) ,  sin(4πx) ,  sin(8πx) .
#¿Que conlusión puedes sacar al respecto?
#Realiza la gráfica con titulo, ejes, etiquetas y todo lo necesario
#para que sea publicable.

x = np.linspace(0, 1, 1000)
y1 = np.sin(2*np.pi*x)
y2 = np.sin(4*np.pi*x)
y3 = np.sin(8*np.pi*x)


with plt.style.context(('ggplot')):
  plt.figure(figsize=(15,7))

  plt.subplot(3, 1, 1)
  plt.plot(x, y1)
  plt.title('2*$\pi$*x')
  plt.ylabel('2*$\pi$*x')
  plt.axis([0, 1, -1.1, 1.1])
  plt.xlabel("x")

  plt.subplot(3, 1, 2)
  plt.plot(x, y2)
  plt.title('4*$\pi$*x')
  plt.ylabel('4*$\pi$*x')
  plt.axis([0, 1, -1.1, 1.1])
  plt.xlabel("x")


  plt.subplot(3, 1, 3)
  plt.plot(x, y3)
  plt.title('8*$\pi$*x')
  plt.ylabel('8*$\pi$*x')
  plt.axis([0, 1, -1.1, 1.1])
  plt.xlabel("x")
  print("Se observa que la frecuencia es mayor")
  plt.show()
```

```python
#Grafíca la función  e−tcos(2πt)  para  t∈[0,5] .
#Asegurate que la gráfica sea una linea punteada de color rojo,
#que la gráfica tenga título,
#etiqueta en el eje de  t  (tiempo), etiqueta en el eje de  y
#(voltaje en  μV ), y una nota donde se escriba la ecuación simulada.

t = np.linspace(0, 5, 100)
y = np.exp(-t) * np.cos(2 * np.pi * t)

with plt.style.context(('ggplot')):
  plt.figure(figsize=(15,7))
  plt.plot(t, y, 'r--')
  plt.axis([-0.1, 5.1, -1.1, 1.1])
  plt.title('e^(-t)*cos(2*pi*t)')
  plt.xlabel('t (tiempo)')
  plt.ylabel('Voltaje en μV')
  plt.annotate('(e^-t)*cos(2*pi*t)', xy=(3, 1))
  plt.show()
```

```python
#Copia un ejemplo de la galería de matplotlib (por ejemplo este o este,
#o este inclusive) y modificalo para que se grafique dentro de la libreta.
#Una vez funcionando, comenta el código, dejando bien claro en español
#y con tus palabras que es lo que hace cada una de las lineas

#SE TOMO LA SIGUIENTE GRÁFICA

fig, ax = plt.subplots() #define ax como se sobreescriben diferentes cosas sobre una misma gráfica
x = np.arange(0, 4 * np.pi, 0.01) # los valores de x de 0 a 4 con incremento de 0.01
y = np.sin(x) #función de x, "y" que es el seno de x
ax.plot(x, y, color='black') #gráfica x vs y en negro

threshold = 0.75 # se define el valor límite
ax.axhline(threshold, color='green', lw=2, alpha=0.7) #hace una linea verde horizontal en 0.75
ax.fill_between(x, 0, 1, where=y > threshold,
                color='green', alpha=0.5, transform=ax.get_xaxis_transform())
#colorea de verde las regiones donde los valores de y son mayores que el límite usando "where".

```
