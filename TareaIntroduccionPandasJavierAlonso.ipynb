```python
#¿Cuál es el menor, el mayor y el precio promedio de la botella
#por país? ¿De que país es la botella de menor precio?
```

```python
df_min= df_wine.groupby('country')['price'].min()
df_max= df_wine.groupby('country')['price'].max()
df_mean= df_wine.groupby('country')['price'].mean()
```

```python
df_merge= pd.merge(df_min, df_max, on='country')
df_merge= pd.merge(df_merge, df_mean, on='country')
df_merge
```

```python
df_merge= df_merge.sort_values('price_x')
df_merge
```

```python
##¿Cuantos vinos hay con aroma a fresa entre otras consideraciones snobs
#que vienen en la descripción?

df_fresas= df_wine[df_wine['description'].str.contains('strawberry','strawberries')]
df_fresas
```

```python
##¿Cuantas designaciones diferentes hay? ¿Cuál es la más repetida?
##¿Cuantas veces se repite?

df_designaciones= df_wine.designation.describe()
df_designaciones
```

```python
##Hacer un dataframe con la variedad, el país y el precio para vinos con un
##costo menor a los $20 dolares.

df_new = df_wine.loc[df_wine.price < 20, ['variety','country','price']]
df_new.sort_values('price')
```
